```python
from network import *
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
people = 1000
connexions = 25000
steps = 3000
```

```python
network = Network(people,connexions)
```

```python
network.simulate(steps)
```

```python
print("memes count " + str(len(network.memes)))
```

```python
data = {'quality' : [x.quality for x in network.memes],
    'shares' : [x.shares for x in network.memes],
    'views' : [x.views for x in network.memes],
    'start' : [x.start for x in network.memes],
    'end' : [x.end for x in network.memes]}
```

```python
df_memes = pd.DataFrame.from_dict(data)
#df_memes.index = df_memes['quality']
df_memes = df_memes[['quality','shares','views','start','end']]
df_memes['lifetime'] =  df_memes['end'] - df_memes['start']
```

```python
df_memes.head()
```

```python
df_finish = df_memes.dropna()
df_finish = df_finish.sort_index()
df_finish
```

```python
df_finish.plot.scatter(x='quality',y='lifetime')
```
